#### b) Indexing and selection of data

Select all students, that have "at_home" as Mjob and "teacher" as Fjob. Additionaly get the studytime of those students.\
Select all students, that have zero failures. Are there more students that have zero failures than those who have at least one?\
Select all students with a uneven ID.\
Select all students that have an even amount of famrel.\
Is there any student, who is not paid, has no activities and no internet?\
\
Use loc, iloc, [] and/or direct attribute access

```python
# 1)
athome_teaching = students.loc[lambda x: (x.Mjob == "at_home") & (x.Fjob == "teacher")].loc[:,"studytime"]
print(f"There are {len(athome_teaching)} students that study at home and work as a teacher.")
```

```python
# 2)
not_failed = students.loc[lambda x: x.failures == 0]
failed = students.loc[lambda x: x.failures > 0]

print(f"{len(failed)} students failed at least once, {len(not_failed)} never failed.")
```

```python
# 3)
id_uneven = students.loc[lambda x: x.ID % 2 != 0]
print(f"There are {len(id_uneven)} students with an uneven ID")
```

```python
# 4)
famrel_even = students.loc[lambda x: x.famrel % 2 == 0]
print(f"There are {len(famrel_even)} students with an even famrel column")
```

```python
# 5) Is there any student, who is not paid, has no activities and no internet?
nonono = students.loc[lambda x: (x.paid == "no") & (x.activities == "no") & (x.internet == "no")]
print(f"There are {len(nonono)} students who are not paid, have no activities and no internet.")
```
