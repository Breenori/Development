## Changing existing methods

```python
# import the necessary packages
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

def predict_2d_for_plotting(x, y, func):
    X, Y = np.meshgrid(x, y)
    Xmat = np.column_stack([X.reshape(X.size), Y.reshape(Y.size)])
    Z = func(Xmat).reshape(x.size, y.size)
    return(X, Y, Z)

def plot_2d_prediction(x, y, func, xval, yval, label=None, figsize=(8, 8), midval=0, cmap_="bwr", svm_model=None):
    X, Y, Z = predict_2d_for_plotting(xval, yval, func)

    plt.figure(figsize=figsize)
    if label and len(label):
        plt.title(label)

    # calculate more levels
    levels = list(range(int(Z.min())*2, int(Z.max())*2))
    print(levels, Z.min(), Z.max())
        
    plt.contourf(X, Y, Z, cmap=cmap_, levels=levels, alpha=0.2)
    
    #paint the margin mid/boundaries in gray
    plt.contour(X, Y, Z, levels=[-1,0,1], colors=["gray","gray","gray"])
    plt.scatter(np.array(x)[:, 0], np.array(x)[:, 1], c=y, cmap=cmap_)
    
    # if a model is given, circle the support vectors
    if svm_model is not None:
        plt.scatter(svm_model.support_vectors_[:, 0], svm_model.support_vectors_[:, 1], s=100, edgecolors="black", facecolors='none', linewidths=1)
    plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split, cross_validate, GridSearchCV
from sklearn.svm import SVC

# Read in the data
data = pd.read_csv(data_dir+"Dataset1.csv")

X = data.iloc[:,:-1]
y = data.iloc[:,-1]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1337)

# Inserting the best parameters from last exercise
model = SVC(kernel="rbf", coef0=16, gamma=2)
model.fit(X,y)
```

```python
plot_2d_prediction(X, y, model.decision_function, 
                   xval=np.linspace(0, 1, 101), yval=np.linspace(0, 1, 101), 
                   svm_model=model, cmap_="coolwarm")
```
